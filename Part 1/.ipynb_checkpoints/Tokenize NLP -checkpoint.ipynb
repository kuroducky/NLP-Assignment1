```python
import os 
import sys
from pathlib import Path
import codecs
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import PorterStemmer
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#Defining the stemming algorithm as porter stemmer
ps = PorterStemmer()

#Providing file directories to necessary platforms for analysis
path_to_instagram = "./NLP/instagram/"
path_to_patents = "./NLP/patents/"
path_to_stackoverflow ="./NLP/stackoverflow/"
```

```python
#Download this if you havent done it before. Otherwise, comment it out
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

### From popular Social Media platform, Instagram

```python
# From instagram 
with os.scandir(path_to_instagram) as entries:
    count = 0 
    for entry in entries:
        if entry.is_file():
            count += 1
            current_file = codecs.open(entry, "r", encoding = "latin-1")
            #This will read the file and print out the output
            txt = current_file.read()
            print("=================================================Normal Text File: " + str(count) + " ===============================================\n")
            print(txt)
            
            print("=================================================Tokenized Words=====================================================\n")
            tokenized_words = word_tokenize(txt)
            print(tokenized_words)
            
            print("==================================================Stemming Words=====================================================\n")
            stemming_words = []
            #Performing stemming on the file and printing out stemming results
            for i in range(len(tokenized_words)):
                stemming_words.append(ps.stem(tokenized_words[i]))
            print(stemming_words)
            print("==================================================End of File : "+str(count)+" ==================================================\n")
```

## Comparing between the number of distinct tokens, and the length distribution of the tokens.

```python
# From instagram 
with os.scandir(path_to_instagram) as entries:
    count = 0 
    pre_stemming_container = []
    post_stemming_container = []
    for entry in entries:
        if entry.is_file():
            count += 1
            current_file = codecs.open(entry, "r", encoding = "latin-1")
            #This will read the file and print out the output
            txt = current_file.read()
            tokenized_words = word_tokenize(txt)
            for i in range(len(tokenized_words)): 
                container = []
                container.append(tokenized_words[i])
                length = len(tokenized_words[i])
                container.append(length)
                pre_stemming_container.append(container)
            
            stemming_words = []
            #Performing stemming on the file appending the stemming results
            for i in range(len(tokenized_words)):
                stemming_words.append(ps.stem(tokenized_words[i]))
   
            for i in range(len(stemming_words)):
                container = []
                container.append(stemming_words[i])
                length = len(stemming_words[i])
                container.append(length)
                post_stemming_container.append(container)
            
```

```python
pre_dict = {
    
}

post_dict = {
    
}

for i in range(len(pre_stemming_container)):
    if pre_stemming_container[i][1] not in pre_dict:
        pre_dict[pre_stemming_container[i][1]] = 1
    else:
        pre_dict[pre_stemming_container[i][1]] += 1
        
for i in range(len(post_stemming_container)):
    if post_stemming_container[i][1] not in post_dict:
        post_dict[post_stemming_container[i][1]] = 1
    else:
        post_dict[post_stemming_container[i][1]] += 1
```

## Plotting the figure

```python
# Plotting for pre-stemming tokens

keys1 = pre_dict.keys()
values1 = pre_dict.values()

keys2 = post_dict.keys()
values2 = post_dict.values()

fig, ax = plt.subplots()

plt.xlabel("Length of Tokens")
plt.ylabel("Number of Tokens")
ax.set_title('Token distribution for Pre-Stemming Tokens')

plt.bar(keys1, values1)
```

```python
fig1, ax1 = plt.subplots()
plt.xlabel("Length of Tokens")
plt.ylabel("Number of Tokens")
ax1.set_title('Token distribution for Post-Stemming Tokens')

plt.bar(keys2, values2, color = 'darkred')
```

```python
fig, ax = plt.subplots()

plt.xlabel("Length of Tokens")
plt.ylabel("Number of Tokens")
ax.set_title('Token distribution for Pre-Stemming Tokens (Blue) and Post-Stemming Tokens (Red)')

plt.bar(keys1, values1)
plt.bar(keys2, values2, color = 'darkred', alpha = 0.7)
```

# Sentence Segmentation

### For Instagram

```python
with os.scandir(path_to_instagram) as entries:
    count = 0
    tokenized_sent_container = []
    three_sentences_instagram = []
    for entry in entries:
        if entry.is_file():
            count += 1
            container = []
            current_file = codecs.open(entry, "r", encoding = "latin-1")
            #This will print out the outputt
            txt = current_file.read()

            tokenized_sentences = sent_tokenize(txt)
            print(tokenized_sentences)
            if count%6 == 0 :
                three_sentences_instagram.append(tokenized_sentences[1])
            for sent in tokenized_sentences:
                result = len(sent.split())
                container.append(result)
            tokenized_sent_container.append(container)
            print("=================" + str(count) + "=============\n")
print(tokenized_sent_container)
```

```python
tokenized_sent_dict = {
    
}

for i in range(len(tokenized_sent_container)):
    for k in range(len(tokenized_sent_container[i])):
        if tokenized_sent_container[i][k] not in tokenized_sent_dict:
            tokenized_sent_dict[tokenized_sent_container[i][k]] = 1
        else:
            tokenized_sent_dict[tokenized_sent_container[i][k]] += 1
print(tokenized_sent_dict)
```

```python
keys_sent = tokenized_sent_dict.keys()
values_sent = tokenized_sent_dict.values()


fig1, ax1 = plt.subplots()
plt.xlabel("Length of Sentences")
plt.ylabel("Number of Sentences with such length")
ax1.set_title('Sentence Distribution after performing Sentence Tokenization')

plt.bar(keys_sent, values_sent, color = 'orange')
```
